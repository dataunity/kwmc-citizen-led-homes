```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("../data/planning/Geocoded_Planning_Applications2010-2014.csv")
```

```python
data_filwood = data[data["GeoAreaLabel"] == "Filwood"]
```

```python
data_filwood
```

```python
status_counts = data_filwood.groupby(['Status']).agg(len)
status_counts.plot(kind='bar', color='red', alpha=1.0, legend=None, title="Filwood Planning Application Status")
```

```python
filwood_vs_easton_counts = data.groupby(['GeoAreaLabel']).agg(len)
filwood_vs_easton_counts.plot(kind='bar', color='red', alpha=1.0, legend=None, title="Planning Applications by area (2010-2014)")
```

```python
filwood_filter = data["GeoAreaLabel"] == "Filwood"
easton_filter = data["GeoAreaLabel"] == "Easton"
# easton_filter
data_easton_filwood = data[filwood_filter | easton_filter]
# data_easton_filwood
filwood_vs_easton_counts = data_easton_filwood.groupby(['GeoAreaLabel']).agg(len)
filwood_vs_easton_counts.plot(kind='bar', color='red', alpha=1.0, legend=None, title="Easton vs Filwood Planning Applications (2010-2014)")
```

```python
decision_counts = data_filwood.groupby(['Decision']).agg(len)
decision_counts.plot(kind='bar', color='red', alpha=1.0, legend=None, title="Filwood Planning Application Decisions")
```

```python
data_easton = data[data["GeoAreaLabel"] == "Easton"]
```

```python
easton_decision_counts = data_easton.groupby(['Decision']).agg(len)
easton_decision_counts.plot(kind='bar', color='red', alpha=1.0, legend=None, title="Easton Planning Application Decisions (2010-2014)")
```

```python
data_clifton = data[data["GeoAreaLabel"] == "Clifton"]
clifton_decision_counts = data_clifton.groupby(['Decision']).agg(len)
clifton_decision_counts.plot(kind='bar', color='red', alpha=1.0, legend=None, title="Clifton Planning Application Decisions (2010-2014)")
```
